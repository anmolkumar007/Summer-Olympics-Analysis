# Summer Olympics Data Analysis Assignment

```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.read_csv('olympics.csv')
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df_new = df[df['Medal']=='Gold']
data = []
for sport in df_new['Sport'].unique():
  data.append([sport,len(df_new[df_new['Sport']==sport])]) 
pd.DataFrame(data,columns=['Sport','Gold medals']).sort_values(by='Gold medals',ascending=False).head().plot(x='Sport',y='Gold medals',kind='bar')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data1 = []
for sport in df['Sport'].unique():
  data1.append([sport,len(df[df['Sport']==sport])])
pd.DataFrame(data1,columns=['Sport','Medals']).sort_values(by='Medals',ascending=False).head().plot(x='Sport',y='Medals',kind='bar')
```

### 4. Which player has won most number of medals? (Top 5)

```python
data2 = []
for ath in df['Athlete'].unique():
  data2.append([ath,len(df[df['Athlete']==ath])])
pd.DataFrame(data2,columns=['Athlete','Medals']).sort_values(by='Medals',ascending=False).head(13).plot(x='Athlete',y='Medals',kind='bar')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df_new[df_new['Country']=='IND'].sort_values(by='Year').head(1)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data4 =[]
for eve in df['Event'].unique():
  data4.append([eve,len(df[df['Event']==eve])])
pd.DataFrame(data4,columns=['Event','No. of Players']).sort_values(by='No. of Players',ascending=False).head().plot(x='Event',y='No. of Players',kind='bar')
```
